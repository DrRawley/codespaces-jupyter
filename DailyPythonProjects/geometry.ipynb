```python
import math

function_definitions = {
  'circle_area': {
    'name': 'Area of Circle',
    'return_value': 'area',
    'formula': 'radius**2 * math.pi',
    'input_number': 1,
    'inputs': ['radius'],
  },
  'circle_circumference': {
    'name': 'Circumference of Circle',
    'return_value': 'circumference',
    'formula': '2 *radius * math.pi',
    'input_number': 1,
    'inputs': ['radius'],
  },
  'rectangle_area': {
    'name': 'Area of Rectangle',
    'return_value': 'area',
    'formula': 'base * height',
    'input_number': 2,
    'inputs': ['base', 'height'],
  },
  'rectangle_perimiter': {
    'name': 'Perimiter of Rectangle',
    'return_value': 'perimiter',
    'formula': '2 * (base + height)',
    'input_number': 2,
    'inputs': ['base', 'height'],
  },
  'triangle_area': {
    'name': 'Area of Triangle',
    'return_value': 'area',
    'formula': '(base * height)/2',
    'input_number': 2,
    'inputs': ['base', 'height'],
  },
  'triangle_perimiter': {
    'name': 'Perimiter of Triangle',
    'return_value': 'perimiter',
    'formula': 'side1 + side2 + side3',
    'input_number': 3,
    'inputs': ['side1', 'side2', 'side3'],
  },
  'sphere_surface_area': {
    'name': 'Surface Area of Sphere',
    'return_value': 'surface area',
    'formula': 'radius**2 * 4 * math.pi',
    'input_number': 1,
    'inputs': ['radius'],
  },
  'sphere_volume': {
    'name': 'Volume of Sphere',
    'return_value': 'volume',
    'formula': 'radius**3 * 4 * math.pi / 3',
    'input_number': 1,
    'inputs': ['radius'],
  },
}

def calculate(function_name, values):
  inputs_string = ''
  for input in function_definitions[function_name]['inputs']:
    inputs_string += input + ', '
  return eval(f"lambda {inputs_string}: {function_definitions[function_name]['formula']}")(*values)

caluculations_to_do = [
  ['circle_area', [1]],
  ['circle_circumference', [1]],
  ['rectangle_area', [2, 3]],
  ['rectangle_perimiter', [2, 3]],
  ['triangle_area', [2, 3]],
  ['triangle_perimiter', [2, 3, 4]],
  ['sphere_surface_area', [1]],
  ['sphere_volume', [1]],
]

for calculation_to_do in caluculations_to_do:
  print(function_definitions[calculation_to_do[0]]['name'], ":", calculate(calculation_to_do[0], calculation_to_do[1]))

```
